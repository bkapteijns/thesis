```python
## Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import re
import os
```

```python
## Obtaining and cleaning the data
train_data = pd.read_csv("titanic_train.csv")
train_data = train_data.drop(["PassengerId", "Name", "Age", "SibSp", "Parch", "Ticket", "Cabin"], axis=1)
X_train = train_data.drop("Survived", axis=1).to_numpy()
col_names = ["Passenger class", "Sex", "Fare", "Embarked"]
y_train = train_data["Survived"].to_numpy()
print(X_train)
```

```python
from sklearn.preprocessing import LabelEncoder

le1 = LabelEncoder()
le3 = LabelEncoder()

le1.fit(X_train[:,1])
le3.fit(X_train[:,3])

X_train[:,1] = le1.transform(X_train[:,1])
X_train[:,3] = le3.transform(X_train[:,3])
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

dtc1 = DecisionTreeClassifier(max_depth=20)
dtc2 = DecisionTreeClassifier(max_depth=2)

dtc1.fit(X_train, y_train)
dtc2.fit(X_train, y_train)
```

```python
plt.figure(figsize=(40, 20))
plot_tree(dtc2, feature_names=col_names, class_names=["survived", "died"], fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(50,50))
plot_tree(dtc1, feature_names=col_names, fontsize=10)
plt.show()
```
