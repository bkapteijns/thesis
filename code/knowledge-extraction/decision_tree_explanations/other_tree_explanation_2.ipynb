```python
## Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import re
import os
```

```python
## Obtaining and cleaning the data
train_data = pd.read_csv("diabetes.csv")
col_names = train_data.columns[:-1]
X_train = train_data.drop("Outcome", axis=1).to_numpy()
y_train = train_data["Outcome"].to_numpy()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

dtc1 = DecisionTreeClassifier(max_depth=20)
dtc2 = DecisionTreeClassifier(max_depth=2)

dtc1.fit(X_train, y_train)
dtc2.fit(X_train, y_train)
```

```python
plt.figure(figsize=(40, 20))
plt.title("Does someone have diabetes?")
plot_tree(dtc2, feature_names=col_names, class_names=["yes", "no"], fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(50,50))
plot_tree(dtc1, feature_names=col_names, fontsize=10)
plt.show()
```
