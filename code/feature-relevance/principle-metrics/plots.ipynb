```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
disaster = pd.read_csv("../SHAP/disaster/axiom_values.csv")
product = pd.read_csv("../SHAP/product-classification/axiom_values.csv")
sentiment = pd.read_csv("../SHAP/sentiment-analysis/axiom_values.csv")
spam = pd.read_csv("../SHAP/spam-classification/axiom_values.csv")
combined = pd.concat([disaster, product, sentiment, spam])
```

```python
for x in ["compactness", "contrast", "distinctiveness", "fidelity", "realism"]:
  print(x)
  value = combined[x]
  plt.boxplot(value)
  plt.show()
  plt.hist(value)
  plt.show()
  print(np.percentile(value, [50, 50]))
  print(np.percentile(value, [40, 60]))
  print("*", np.percentile(value, [33, 67]))
  print("-", np.percentile(value, [25, 75]))
  print(np.percentile(value, [20, 80]))
  print(np.percentile(value, [10, 90]))
  print(np.percentile(value, [5 , 95]))
  print(np.percentile(value, [0 ,100]))
```
